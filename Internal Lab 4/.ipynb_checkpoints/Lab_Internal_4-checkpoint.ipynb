# Q5 Remove any unwanted columns or outliers, standardize variables in pre-processing step

```python
# We will consider all columns for outlier analysis except ID and Class as ID is not valid for analysis and Class is
# our target variable. 
capDf = pd.DataFrame()
toCapDf = df.loc[:,:]
```

```python
for col in df.columns:
    if(col == 'Class' or col == 'ID'):
        capDf[col] = df[col]
        continue
    if(col in toCapDf.columns):
        percentiles = toCapDf[col].quantile([0.25,0.75]).values
        Q1 = percentiles[0]
        Q3 = percentiles[1]
        IQR = Q3 - Q1
        minCap = Q1 - (IQR) * 1.5
        maxCap = Q3 + (IQR) * 1.5
        capDf[col] = toCapDf[col][(toCapDf[col] >= minCap) & (toCapDf[col] <= maxCap)]
    else:
        capDf[col] = df[col]
    
capDf.dropna(inplace=True);
```

```python
#Printing the outliers which were removed (In all, 191 eligible outlier rows were removed)
outlierDf = pd.concat([df, capDf])
outlierDf = outlierDf.reset_index(drop=True)
df_gpby = outlierDf.groupby(list(outlierDf.columns))
idx = [x[0] for x in df_gpby.groups.values() if len(x) == 1]
outlierDf.reindex(idx)
```

```python
#Standardization
from sklearn import preprocessing
X = capDf.loc[:, df.columns != 'Class']
Y = capDf[['Class']]
x_scaled = preprocessing.StandardScaler().fit_transform(X)
X_std = pd.DataFrame(x_scaled, columns = X.columns)
```
