```python
import numpy as np
import pandas as pd
import io
```

```python
df = pd.read_csv('Laliga.csv')
df.columns = df.iloc[0]
df.drop(df.index[[0]], inplace=True)
df.replace(['-'],['0'], inplace=True)
df
```

```python
df[(df['Debut'] >= '1930') & (df['Debut'] <= '1980')][['Team','Debut']]
```

```python
df[['Points']] = df[['Points']].apply(pd.to_numeric)
df.sort_values('Points', ascending=False, inplace=True)
df[['Team','Points']].head(5)
```

```python
def Goal_diff_count(row):
    return row['GoalsFor'] - row['GoalsAgainst']
```

```python
df[['GoalsFor']] = df[['GoalsFor']].apply(pd.to_numeric)
df[['GoalsAgainst']] = df[['GoalsAgainst']].apply(pd.to_numeric)
df['GoalDifference'] = df.apply(Goal_diff_count, axis=1)
df[['Team', 'GoalDifference']]
```

```python
maxDiff = df['GoalDifference'].max()
minDiff = df['GoalDifference'].min()
df[(df['GoalDifference'] == maxDiff) | (df['GoalDifference'] == minDiff)][['Team','GoalDifference']]
```

```python
def WinningPercent(row):
    return (row['GamesWon']/row['GamesPlayed'])*100
```

```python
df[['GamesWon']] = df[['GamesWon']].apply(pd.to_numeric)
df[['GamesPlayed']] = df[['GamesPlayed']].apply(pd.to_numeric)
df[['GamesWon']] = df[df['GamesWon'] != 0]['GamesWon']
df[['GamesPlayed']] = df[df['GamesPlayed'] != 0]['GamesPlayed']
df['Winning Percent'] = df.apply(WinningPercent, axis = 1)
df['Winning Percent'].fillna(0, inplace=True)
df
```

```python
df.sort_values('Winning Percent', ascending = False, inplace=True)
df.head(5)
```
